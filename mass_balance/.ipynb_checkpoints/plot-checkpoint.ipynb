```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('rgi_mass_balance.csv')
data
```

```python
#volume_now = data['volume']
#volume_then = data['volume'] - data['volume_change']
plt.rcParams['figure.figsize'] = [16, 12]
avg_H0 = (data['volume'] - data['volume_change']) / data['area']
avg_H1 = data['volume'] / data['area']

#area = data['area'] / 1e6
#area[area > 100] = 100.

plt.scatter(avg_H0, avg_H1, s = 3, alpha = 1, c=data['area'] / 1e6, vmin = 0., vmax=25.)
cbar = plt.colorbar()
cbar.set_label(r'Glacier Area (km$^2$)', rotation=270)
plt.plot([0.,500], [0.,500.], 'ro-')
plt.xlim([0.,500.])
plt.ylim([0.,500.])
plt.xlabel('Avg. Thickness 2000 (m)')
plt.ylabel('Avg. Thickness 2019? (m)')
```

```python
plt.scatter(avg_H0, avg_H1, s = 3, alpha = 1, c=data['area'] / 1e6, vmin = 0., vmax=2.5)
cbar = plt.colorbar()
cbar.set_label(r'Glacier Area (km$^2$)', rotation=270)
plt.plot([0.,500], [0.,500.], 'ro-')
plt.xlim([0.,100.])
plt.ylim([0.,100.])
plt.xlabel('Avg. Thickness 2000 (m)')
plt.ylabel('Avg. Thickness 2019? (m)')
```
