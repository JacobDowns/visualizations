```python
import numpy as np
import pandas as pd
import json

df = pd.read_csv('plot_data.csv', dtype={"rgi_index": str})
f = open('plot_data.geojson')
data = json.load(f)
import plotly.express as px

fig = px.choropleth_mapbox(df, geojson=data, 
                           locations='rgi_index', 
                           featureidkey = 'properties.rgi_index',
                           color='avg_dh',
                           color_continuous_scale="Viridis_r",
                           mapbox_style="carto-positron",
                           zoom=5,
                           center = {"lat": 60.42, "lon": -142.58},
                           opacity=0.5,
                          )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
